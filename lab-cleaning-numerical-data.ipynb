### 7. Define a function that differentiates between continuous and discrete variables. Hint: The number of unique values might be useful. Store continuous data into a continuous_df variable and do the same for discrete_df

```python
numerical_df.describe()
```

```python
numerical_df.nunique()
```

You can see that most numerical variables appear to be continuous.

However, number_of_open_complaints and number_of_policies would count as discrete as there are so few unique values.

```python
def cont_disc(df):
    
    col = []
    
    for i in df.columns:
        if df[i].nunique() < 10: # arbitrarily saying 10 is the cut-off
            col.append(i)
    
    continuous = df.drop(col, axis=1)
    discrete = df[col]
    
    return continuous, discrete
```

```python
continuous_df, discrete_df = cont_disc(numerical_df)
display(continuous_df)
discrete_df
```

```python
# from Amr

def split_continuous_discrete(df):
    
    discrete_columns = [col for col in df.columns if df[col].nunique() <= 10]
    
    continuous_columns = [col for col in df.columns if df[col].nunique() > 10]

    return df[continuous_columns], df[discrete_columns]
```

### 8. Compare policy_type and policy. What information is contained in these columns. Can you identify what is important?

```python
categorical_df.groupby(['policy_type','policy']).agg({'customer':pd.Series.nunique})
```

Information is repeated between the two columns so something can be removed.

Option 1 - we could remove feature policy_type which would ultimately lead to 8 columns generated by OneHot encoding (drop first).

Option 2 - we could group policies into 1, 2 and 3 and leave policy_group. This would ultimately lead to 4 columns generated by OneHot encoding (drop first).

I'll go with option 2.

```python
categorical_df['policy_type'] = categorical_df['policy_type'].replace({'Corporate Auto':'C','Personal Auto':'P','Special Auto':'S'})

categorical_df['policy'] = np.where(categorical_df['policy'].isin(['C1','P1','S1']) , '1', categorical_df['policy'])
categorical_df['policy'] = np.where(categorical_df['policy'].isin(['C2','P2','S2']) , '2', categorical_df['policy'])
categorical_df['policy'] = np.where(categorical_df['policy'].isin(['C3','P3','S3']) , '3', categorical_df['policy'])
categorical_df[['policy','policy_type']]
```

### 9. Check number of unique values in each column, can they be combined in any way to ease encoding? Comment your thoughts and make those changes.

```python
categorical_df.nunique()
```

Consolidate education values Bachelor, Master and Doctor into a new value University.

Consolidate lower frequency employmentstatus values into Other.

Consolidate lower frequency vehicle_class  values into Luxury.

```python
categorical_df['education'] = np.where(categorical_df['education'].isin(['Bachelor','Master','Doctor']) , 'University', categorical_df['education'])

categorical_df['employmentstatus'] = np.where(categorical_df['employmentstatus'].isin(['Disabled','Retired','Medical Leave']) , 'Other', categorical_df['employmentstatus'])

categorical_df['vehicle_class'] = np.where(categorical_df['vehicle_class'].isin(['Sports Car','Luxury SUV','Luxury Car']) , 'Luxury', categorical_df['vehicle_class'])

print(categorical_df['education'].value_counts())
print(categorical_df['employmentstatus'].value_counts())
print(categorical_df['vehicle_class'].value_counts())
```
